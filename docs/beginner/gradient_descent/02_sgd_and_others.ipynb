# From Gradient Descent to SGD

Statt mit einer Target Funktion aus synthetischem Daten zu arbeiten, schauen wir uns die Wirkung verschiedener Optimizer auf die Loss Funktion an. Statt unsere Prediction wie vorher zu optimieren, um ein geringes MSE zu erreichen, optimieren wir nun die Eingabeparameter der Loss Funktion. Es wird kein Modell zuf√§llig initalisert, hier geht es darum, die Parameter der Loss Funktion mit verschiedenen Verfahren zu optimieren.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
def himmelblau(x,y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

def mesh(start,end,num=500):
    x = np.linspace(start,end,num)
    y = np.linspace(start,end,num)
    return np.meshgrid(x,y)

pick_function = {
    "himmelblau": (himmelblau, mesh(-6,6))
}
```

```python
## Visualisierung der Landscapes
himmelblau, (x1,y1) = pick_function["himmelblau"]
z1 = himmelblau(x1,y1)
```

```python
fig1, ax1 = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax1.plot_surface(x1, y1, z1,
                       cmap=cm.turbo,
                       vmax=np.percentile(z1, 95),
                       linewidth=0, antialiased=True,
                       #rstride=5, cstride=5,
                       zsort='average')
ax1.view_init(elev=50, azim=-20)
```
